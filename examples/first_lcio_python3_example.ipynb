```python
from ROOT import TH1D, TCanvas, TF1, std
from pyLCIO import EVENT, UTIL, IOIMPL
import sys
from math import *
```

```python
fName="/home/ilc/LCIO/examples/cpp/rootDict/DST01-06_ppr004_bbcsdu.slcio"

reader = IOIMPL.LCFactory.getInstance().createLCReader()
reader.open( fName )
```

```python
# print the collection names of the first 10 events

maxEvt = 10
nEvt = 0

for evt in reader:
    nEvt += 1
    if nEvt > maxEvt:
            break

    colNames = evt.getCollectionNames()
    print("--------------------")
    for n in colNames:
        print("  ", n)
    print("---------------------")
```

```python
# ---- create a histogram with pfo energies
hEpfo   = TH1D( 'Epfo',   'pfo energy',50, 0., 50. )


maxEvt = 10
nEvt = 0

for evt in reader:
    nEvt += 1
    if nEvt > maxEvt:
            break

    pfos = evt.getCollection("PandoraPFOs")
    
    for pfo in pfos:
        hEpfo.Fill( pfo.getEnergy() )
       #print( " filling energy: ", pfo.getEnergy() )

c1 = TCanvas( "plots" , "plots" , 210*2 , 297*3 )        
hEpfo.Draw()
c1.Draw()
#userInput = input( 'Press any key to continue' )
```
